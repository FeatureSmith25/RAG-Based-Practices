**Pydantic Tool**

```python
from langchain_community.tools import StructuredTool
from pydantic import BaseModel, Field
```

```python
class mathinput(BaseModel):
    a:int=Field(required=True, description="The first numer to add")
    b:int=Field(required=True, description="The second numer to add")
```

```python
def multiply_func(a:int, b:int)-> int:
    return a*b;
```

```python
multiply_tool=StructuredTool.from_function(
    func=multiply_func,
    name="multiply",
    description="Multiply two numbers",
    args_schema=mathinput
)
```

```python
result=multiply_tool.invoke({'a':3, 'b':3})
print(result)
```

**Base Tool**

```python
from langchain_community.tools import BaseTool
from typing import Type
```

```python
class multiplytool(BaseTool):
    name: str="multiply"
    description: str="Multiply two numbers"
    args_schema: Type[BaseModel]=mathinput
    def _run(self, a:int, b:int)-> int:
        return a*b
```

```python
class addtool(BaseTool):
    name:str="add"
    description: str="add two numbers"
    args_schema:Type[BaseModel]=mathinput
    def _run(self, a:int, b:int)->int:
        return a+b
```

```python
class MathToolKit:
    def get_tools(self):
        return [addtool,multiplytool]
```

```python
toolkit=MathToolKit()
tools=toolkit.get_tools()

for tool in tools:
    print(tool().invoke({"a": 3, "b": 4}))
```
