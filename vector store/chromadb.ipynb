```python
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import Chroma
```

```python
from langchain_core.documents import Document
```

```python
doc1 = Document(
    page_content="Shubman Gill is an exceptionally talented top-order batsman known for his elegant strokeplay and consistent run-scoring at the top of the order for Gujarat Titans.",
    metadata={"team": "Gujarat Titans"}
)

doc2 = Document(
    page_content="Rishabh Pant is an explosive wicketkeeper-batsman celebrated for his aggressive, unconventional batting style and dynamic match-winning abilities for Delhi Capitals.",
    metadata={"team": "Delhi Capitals"}
)

doc3 = Document(
    page_content="Sanju Samson is a highly skilled wicketkeeper-batsman and the influential leader of Rajasthan Royals, famous for his effortless six-hitting and timing.",
    metadata={"team": "Rajasthan Royals"}
)

doc4 = Document(
    page_content="Shreyas Iyer is a composed middle-order batsman and a proven captain who has successfully anchored innings and led Kolkata Knight Riders with strategic acumen.",
    metadata={"team": "Kolkata Knight Riders"}
)

doc5 = Document(
    page_content="KL Rahul is a versatile and technically sound opening batsman who has consistently ranked among the top run-scorers in the tournament while leading Lucknow Super Giants.",
    metadata={"team": "Lucknow Super Giants"}
)
```

```python
docs=[doc1, doc2, doc3, doc4, doc5]
```

```python
vector_store=Chroma(
    embedding_function=HuggingFaceEmbeddings(),
    persist_directory='chroma_db',
    collection_name='sample'
)
```

```python
vector_store.add_documents(docs)
```

```python
vector_store.get(include=['embeddings', 'documents', 'metadatas'])
```

```python
# Searching Documents
vector_store.similarity_search(
    query="who among these are a batsman",
    k=2
)
```

```python
# searching documents
vector_store.similarity_search_with_score(
    query="who among these are a batsman",
    k=2
)
```

```python
# searching with filter
vector_store.similarity_search_with_score(
    query="",
    filter={"team": "Kolkata Knight Riders"}
)
```

```python
updated_doc1=Document(
    page_content="KL Rahul is a versatile and technically sound opening batsman who has consistently ranked among the top run-scorers in the tournament. Known for his elegant strokeplay and leadership, he serves as an influential captain anchoring the top-order while doubling as an elite wicketkeeper.",
    metadata={"team": "Lucknow Super Giants"}
)
vector_store.update_document(document_id='bbc14acf-c3ca-4cf0-a105-1cdaef7ef8c5', document=updated_doc1)
```
